### 1. Convert the `weekday` column from a number to a string (1 point)

Knowing that 0 represents Monday, 1 represents Tuesday, etc., convert the `weekday` column from a number to a string. 

Use `map` to perform this operation. You can mix it with a `lambda` function.

##### Explanation
First, we create a dictionary consisting of integers **0** to **6** and the names of the weekdays.
Second, we use the dictionary in a *lambda* function to override the integer values with the strings defined in the dictionary.

```python
#Dictionary containing the weekdays
dict_week_day_name = {
    0: "Monday", 
    1: "Tuesday", 
    2: "Wednesday", 
    3: "Thursday", 
    4: "Friday", 
    5: "Saturday", 
    6: "Sunday"
}

#Replace of the weekdays number for it's name.
energy["weekday"]=energy["weekday"].map(lambda d: dict_week_day_name[d])

#We obtain a sample of the data to validate our changes
energy.sample(10, random_state = 1)
```

### 4. How many hours per month, in average, do we have a price above the monthly average? And below? (1 point)

##### Explanation
First, we create a dataframe called "*price_daily*", which we use to store only the column month and *spot_price* and another dataframe called "avg_price_month", where we save the average spot price per month. 

By using the column "*month*" as the index in both dataframes, we can now divide the dataframe "*price_daily*" by "*avg_price_month*". If the value in the column "*spot_price*" in the resulting dataframe "*df_month*" is higher than 1, the spot price is above the monthly average. If the value in the column "*spot_price*" in the resulting dataframe "*df_month*" is below 1, the spot price is below the monthly average. Please note that, for reasons of simplicity, we assign the category "*below*" to all spot prices that are equal to 1 as the question focuses on the ones above the monthly average. 

We create a summary table per month by summing up the values for the columns "*above*","*below*" and "*total*" (which is the sum of "*above*" and "*below*") and use this dataframe to create the final dataframe called "*df*", where we show a summary for each month and append a line called "*TOTAL SUM*" with the yearly average of hours per month that are above the monthly average, which we highlight in blue. 

```python
#Creation of the dataframes
price_daily = energy[["month","spot_price"]]
price_daily = price_daily.set_index("month")

#Obtaing the statistics 
avg_price_month = energy.groupby(['month'])[['spot_price']].mean()
df_month = price_daily/avg_price_month
df_month['above'] = df_month['spot_price'] > 1
df_month['below'] = df_month['spot_price'] <= 1
df_month['total'] = df_month['above'] + df_month['below']
df_count=df_month.groupby('month')[['above', 'below', 'total']].sum()

#Creating the summary 
df=pd.DataFrame({
    "Hours above average":df_count["above"],
    "Monthly average above":df_count["above"]/12,
    "Hours below average":df_count["below"],
    "Monthly average below":df_count["below"]/12,
})

#We add the percentage of the monthly average and the total sum
new_row = pd.Series(data={"Hours above average":int(df_count["above"].sum()),"Monthly average above":int(df_count["above"].sum()/12),
                          "Hours below average":int(df_count["below"].sum()), "Monthly average below":int(df_count["below"].sum()/12)},
                            name='TOTAL SUM')

#
df = df.append(new_row, ignore_index=False)
df.style.highlight_max(["Monthly average above","Monthly average below"],color='blue')
```

* Is the `spot_price` higher or lower than its hourly average when `solar` generation is above hourly its average?
* Is the `spot_price` higher or lower than its monthly average when `power_demand` is above its monthly average?


##### Explanation
For the hourly evaluation, we calculate the hourly average of gas and wind. The same applies to spot price and solar, which is therefore not further being explained. 

We use a nested for loop to iterate through all possible **24 hours (0-23)** and all rows in the dataframe "*energy*". If for one row the wind production is above its hourly average, we add 1 to a variable called "*wind_above*". If also the gas production is above its hourly average, we add 1 to a variable called "*gas_above*". By dividing the final value of "*gas_above*" by the final value of "*wind_above*" we get the percentage of the number of rows where the gas generation is higher than its hourly average when the wind is also higher than its hourly average.

For the last question, we apply the same concept with the difference that we iterate through the dictionary "*dict_month_name*" from question 2 in order to loop through all possible 12 months.

```python
#We obtain per hour mean from gas and wind
hourly_avg_gas=energy.groupby("hour")["gas"].mean()
hourly_avg_wind=energy.groupby("hour")["wind"].mean()

#For loop to obtain the amount of times gas goes above it's average when wind goes above it's average
gas_above=0
wind_above=0
for hour in range(len(hourly_avg_gas)):
    for x in range(len(energy)):
        if (hour==energy["hour"][x]) and (energy["wind"][x]>hourly_avg_wind[hour]):
            wind_above+=1
            if (energy["gas"][x]>hourly_avg_gas[hour]):
                gas_above+=1
higher_percentage=format(gas_above/wind_above,"%")

#Presenting the resutls
print("The total cases where wind generation is above its hourly average are {}. And the total cases where gas generation is higher than its hourly average in the cases where wind generation is also above its hourly average are {}. So we can conclude saying that {} of the cases the gas generation is higher than its hourly average when the wind is also higher. In conclusion, gas generation is lower than its hourly average when wind generation is above its hourly average.".format(wind_above,gas_above,higher_percentage))
```

```python
#We obtain per hour mean from spot_price and solar
hourly_avg_spot_price=energy.groupby("hour")["spot_price"].mean()
hourly_avg_solar=energy.groupby("hour")["solar"].mean()

#For loop to obtain the amount of times solar goes above it's average when spot_price goes above it's average
spot_price_above=0
solar_above=0

Total_spot_pirce = 0
for hour in range(len(hourly_avg_spot_price)):
    for x in range(len(energy)):
        if (hour==energy["hour"][x]) and (energy["solar"][x]>hourly_avg_solar[hour]):
            solar_above+=1
            Total_spot_pirce -= energy["spot_price"][x]
            if (energy["spot_price"][x]>hourly_avg_spot_price[hour]):
                spot_price_above+=1
                Total_spot_pirce += energy["spot_price"][x]
higher_percentage=format(spot_price_above/solar_above,"%")

#Presenting the resutls
print("The total cases where solar generation is above its hourly average are {}. And the total cases where the spot price is higher than its hourly average in the cases where solar generation is also above its hourly average are {}, so we can conclude saying that half of the time ({} of the cases) the spot price is higher than its hourly average when the solar generation is also higher.".format(solar_above,spot_price_above,higher_percentage))
```

### 8. Weekends vs weekdays (1 point)

Create a column called `weekend` that is `True` if the day is a weekend and `False` if it's a weekday. Use `np.where` to create this column.

How much expensive in average is a weekend day compared to a weekday?

##### Explanation
We add a column called "*weekends*" to the dataframe "*energy*", which we populate with "*True*" if the value in column "*weekday*" is a Saturday or Sunday. If not, we assign "*False*". 

Then, we calculate the average spot price for all rows with the value "*True*" in column "*weekends*" and deduct the average spot price of all rows with the value "*False*" in column "*weekends*". The resulting value is the difference of the average spot price during weekends and during weekdays.

```python
#Generating a boolean for the days of the week.
energy["weekends"] = np.where(
    energy["weekday"].isin(["Saturday","Sunday"]), 
    True, 
    False 
)

#Generating the average per category
weekends_avg=energy.groupby("weekends")["spot_price"].mean()[True]
weekdays_avg=energy.groupby("weekends")["spot_price"].mean()[False]
difference=weekends_avg-weekdays_avg

#Cheking h the difrence and their prices.
if difference>0:
    print("""The average price of the weekends is {:.2f}â‚¬ and the average price of the weekdays is {:.2f}â‚¬.
    Weekends are {:.2f}â‚¬ more expensive than weekdays in average.""".format(weekends_avg,weekdays_avg,difference))
elif difference<0:
    print("""The average price of the weekends is {:.2f}â‚¬ and the average price of the weekdays is {:.2f}â‚¬.
    Weekends are {:.2f}â‚¬ cheaper than weekdays in average.""".format(weekends_avg,weekdays_avg,abs(difference)))
else:
    print("""The average price of the weekends is {:.2f}â‚¬ and the average price of the weekdays is {:.2f}â‚¬.
    In average the price for weekends and weekdays is the same.""".format(weekends_avg,weekdays_avg))
```

Spoiler: you can plot in `pandas` with `df.plot()`.

##### Explanation
For the open questions, we use the Python libraries "*Pandas*" for the dataframe handling, "*NumPy*" for a linear regression and "*matplotlib.pyplot*" for plotting the graphs. 

First, we create a correlation matrix of the columns of the dataframe "*energy*" to get an overview of how the different variables, i.e., columns, are correlated. We use the correlation matrix to obtain the correlation of each energy source with the spot price.

Second, we create a scatter plot for each energy source abd spot price. We show the energy source on the x-axis and the spot price on the y-axis. We do a linear regression on these two parameters, which we add to the scatter plot. 

The regression is done by using the function "*polyfit*", available in NumPy. The first two arguments are the two varibles x and y, e.g., "*hydro*" and "*spot_price*" in one of our cases. By setting the degree to **1**, we define that the output is a linear regression (and not a funtion of a higher degree like a parabola). The linear regression is a function of the type: y = ax+b, where a is the slope and b is the intersect with the y-axis.

We create the new column "*regression*", which we add to the dataframe "*energy*". In this column we store the output (y) of the linear regression function when using the energy source as input (x).

We create the scatter plot and the regression and print the linear regression. 

All steps involved in doing the regression and plotting the graphs are combined in a function, which we call "*regression*". By using a function, we can plot the graphs for each data source without repeating the code. 



```python
#Importing matplotlib for the graphics generations 
import matplotlib.pyplot as plt
```

```python
#Correlation Matrix
corr= energy[["spot_price","nuclear","gas","solar","wind","hydro","coal"]].corr()
corr.style.background_gradient (cmap = "PuBu_r")
```

```python
#Defining fucntion to generate the regression graphs.
def regression(str1,str2):
    #calculate parameters for linear regression
    poly=np.polynomial.polynomial.polyfit(energy[str1],energy[str2],deg=1) 
    
    #create regression column
    energy["regression"]=poly[0]+poly[1]*energy[str1] #create regression column
    
    #create scatter plot
    ax=energy.plot(x=str1, y=str2, kind="scatter",figsize=(20,10),alpha=0.5) 
    
    #plot regression
    energy.plot(x=str1,y="regression",color="Red",ax=ax) 
    ax.text(energy[str1].min(),-9,'linear regression: Y = {:.4f} + {:.4f}X'.format(poly[0],poly[1]), fontsize=15) #print funcion
    plt.xlabel(str1,fontsize=15)
    plt.ylabel(str2,fontsize=15)
    plt.title(label=str1.upper() +" vs. "+str2.upper(),fontsize=20,fontweight="bold")
```
